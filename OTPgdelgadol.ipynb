Algoritmo de One Time Pad que utiliza el módulo 26, o sea, mensajes en inglés y no se toman en cuenta caracteres especiales, números ni espacios.

La función `keyGenerator` toma como parámetro el mensaje que se desea encriptar y crea una llave de la longitud del mensaje limpio, o sea, sin caracteres especiales, números ni espacios. Además, todos los caracteres se encuentran en mayúscula.

```python
import random
import re

def keyGenerator(plaintext):
  cleanedPlaintext = re.sub(r'[^a-zA-Z]', '', plaintext.upper())
  key = []
  for i in range(len(cleanedPlaintext)):
    key.append(random.randint(0, 25))
  return key

def encryptionOTPMod26(plaintext, key):
  cleanedPlaintext = re.sub(r'[^a-zA-Z]', '', plaintext.upper())
  charDict = {chr(65 + i): i for i in range(26)}
  numDict = {i: chr(65 + i) for i in range(26)}
  ciphertext = ""
  for i in range(len(cleanedPlaintext)):
    cipherLetter = (charDict[cleanedPlaintext[i]] + key[i]) % 26
    ciphertext = ciphertext + numDict[cipherLetter]
  return ciphertext

def descryptionOTPMod26(ciphertext, key):
  charDict = {chr(65 + i): i for i in range(26)}
  numDict = {i: chr(65 + i) for i in range(26)}
  plaintext = ""
  for i in range(len(ciphertext)):
    plainLetter = (charDict[ciphertext[i]] - key[i] + 26) % 26
    plaintext = plaintext + numDict[plainLetter]
  return plaintext

message = input()
key = keyGenerator(message)
print("La llave generada: ", key)
cipher = encryptionOTPMod26(message, key)
print("Mensaje cifrado: ", cipher)
plain = descryptionOTPMod26(cipher, key)
print("Mensaje descifrado: ", plain)


```
