```python
import re

def vigenereEncryption(plaintext, key, t):
  cleanedPlaintext = re.sub(r'[^a-zA-Z]', '', plaintext.upper())
  cleanedKey = re.sub(r'[^a-zA-Z]', '', key.upper())
  charDict = {chr(65 + i): i for i in range(26)}
  numDict = {i: chr(65 + i) for i in range(26)}
  ciphertext = ""
  for i in range(len(cleanedPlaintext)):
    ciphertext += numDict[(charDict[cleanedPlaintext[i]] + charDict[cleanedKey[i%(len(key))]])%26]
  ciphertext_spaced = re.sub(r'(.{'+str(t)+'})', r'\1 ', ciphertext)
  return ciphertext_spaced

def vigenereDecryption(ciphertext, key, t):
  cleanedCiphertext = re.sub(r'[^a-zA-Z]', '', ciphertext.upper())
  cleanedKey = re.sub(r'[^a-zA-Z]', '', key.upper())
  charDict = {chr(65 + i): i for i in range(26)}
  numDict = {i: chr(65 + i) for i in range(26)}
  plaintext = ""
  for i in range(len(cleanedCiphertext)):
    plaintext += numDict[(charDict[cleanedCiphertext[i]] - charDict[cleanedKey[i%(len(key))]])%26]
  plaintext_spaced = re.sub(r'(.{'+str(t)+'})', r'\1 ', plaintext)
  return plaintext_spaced

while True:
  flag = input("Ingrese 0 si desea encriptar, 1 si desea desencriptar, 2 si desea salir: ")
  match flag:
    case "0":
      message = input("Ingrese el texto a encriptar: ")
      key = input("Ingrese la clave: ")
      t = int(input("Ingrese el tamaño de los bloques: "))
      ciphertext = vigenereEncryption(message, key, t)
      print("Mensaje encriptado: ", ciphertext)
      print("-------------------------------------------------------------------")
    case "1":
      message = input("Ingrese el texto a desencriptar: ")
      key = input("Ingrese la clave: ")
      t = int(input("Ingrese el tamaño de los bloques: "))
      plaintext = vigenereDecryption(message, key, t)
      print("Mensaje desencriptado: ", plaintext)
      print("-------------------------------------------------------------------")
    case "2":
      print("Chao :)")
      break
    case _ :
      print("Ingrese una opcion valida")
      print("-------------------------------------------------------------------")
```
