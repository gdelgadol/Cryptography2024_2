Nota: En la función de desencriptación se verifica que la entrada del texto cifrado sea par ya que es imposible que en esta implementación se tenga un texto cifrado impar debido a la naturaleza de lo bloques. Adicionalmente, en el caso de que se ingresen valores en la matriz que sean mayor o igual a 26, se informa que no es válido y que deben estar entro 0 y 25, inclusivo.

```python
import numpy as np
import re

def modularInverse(a):
    for i in range(1,26):
      if (i*a)%26==1:
        return i
    return None

def invertMatrix(matrix):
    det = int(np.linalg.det(matrix))
    det_modular_inverse = modularInverse(det)
    if not(det_modular_inverse) or det == 0:
      return None
    matrix[0][0], matrix[1][1] = ((matrix[1][1]*det_modular_inverse) % 26), ((matrix[0][0]*det_modular_inverse) % 26)
    matrix[0][1] = (-1 * matrix[0][1] * det_modular_inverse) % 26
    matrix[1][0] = (-1 * matrix[1][0] * det_modular_inverse) % 26
    return matrix

def hillEncryption(plaintext, key):
    if modularInverse(int(np.linalg.det(key))) == None:
        return None
    cleanedPlaintext = re.sub(r'[^a-zA-Z]', '', plaintext.upper())
    charDict = {chr(65 + i): i for i in range(26)}
    numDict = {i: chr(65 + i) for i in range(26)}
    if len(cleanedPlaintext) % 2 != 0:
        cleanedPlaintext += 'X'
    ciphertext = ""
    for i in range(0, len(cleanedPlaintext), 2):
        pair = np.array([charDict[cleanedPlaintext[i]], charDict[cleanedPlaintext[i+1]]])
        cipherPair = np.dot(pair, key) % 26
        ciphertext += numDict[cipherPair[0]] + numDict[cipherPair[1]]
    return ciphertext

def hillDecryption(ciphertext, key):
    if modularInverse(int(np.linalg.det(key))) == None:
        return None
    cleanedCiphertext = re.sub(r'[^a-zA-Z]', '', ciphertext.upper())
    if len(cleanedCiphertext) % 2 != 0:
        return -1
    keyInverted = invertMatrix(key)
    charDict = {chr(65 + i): i for i in range(26)}
    numDict = {i: chr(65 + i) for i in range(26)}
    plaintext = ""
    for i in range(0, len(cleanedCiphertext), 2):
        pair = np.array([charDict[cleanedCiphertext[i]], charDict[cleanedCiphertext[i+1]]])
        plainPair = np.dot(pair, keyInverted) % 26
        plaintext += numDict[plainPair[0]] + numDict[plainPair[1]]
    return plaintext

while True:
  flag = input("Ingrese 0 si desea encriptar, 1 si desea desencriptar, 2 si desea salir: ")
  match flag:
    case "0":
      message = input("Ingrese el texto a encriptar: ")
      print("Ingrese una matrix 2x2 invertible:")
      print("Ejemplo:")
      print("|1 2|")
      print("|3 4|")
      print("Entrada: 1 2 3 4")
      print("-----------------")
      while True:
          try:
              k1, k2, k3, k4 = map(int, input("Ingrese los datos separados por un espacio: ").split())
              if all(0 <= k < 26 for k in [k1, k2, k3, k4]):
                  break
              else:
                  print("Error: Todos los valores de k deben estar entre 0 y 25. Inténtelo de nuevo.")
          except ValueError:
              print("Error: Ingrese solo números enteros. Inténtelo de nuevo.")
      key = np.array([[int(k1), int(k2)], [int(k3), int(k4)]])
      ciphertext = hillEncryption(message, key)
      if not(ciphertext):
        print("Esta llave no tiene inversa, inténtelo de nuevo.")
      else:
        print("Mensaje encriptado: ", ciphertext)
      print("-------------------------------------------------------------------")
    case "1":
      message = input("Ingrese el texto a desencriptar: ")
      print("Ingrese una matrix 2x2 invertible:")
      print("Ejemplo:")
      print("|1 2|")
      print("|3 4|")
      print("Entrada: 1 2 3 4")
      print("-----------------")
      while True:
          try:
              k1, k2, k3, k4 = map(int, input("Ingrese los datos separados por un espacio: ").split())
              if all(0 <= k < 26 for k in [k1, k2, k3, k4]):
                  break
              else:
                  print("Error: Todos los valores de k deben estar entre 0 y 25. Inténtelo de nuevo.")
          except ValueError:
              print("Error: Ingrese solo números enteros. Inténtelo de nuevo.")
      key = np.array([[int(k1)%26, int(k2)%26], [int(k3)%26, int(k4)%26]])
      plaintext = hillDecryption(message, key)
      match plaintext:
        case -1:
          print("El texto cifrado debe ser de longitud par para ser válido.")
        case None:
          print("Esta llave no tiene inversa, inténtelo de nuevo.")
        case _:
          print("Mensaje desencriptado: ", plaintext)
      print("-------------------------------------------------------------------")
    case "2":
      print("Chao :)")
      break
    case _ :
      print("Ingrese una opcion valida")
      print("-------------------------------------------------------------------")



```
